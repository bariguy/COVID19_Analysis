## Main Goal with OWID Datasets
- Do anoter set of analysis analagous to OWID analysis with US-focused time series datasets

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
time_series_covid19_confirmed_US = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv"
df_time_series_covid19_confirmed_US = pd.read_csv(time_series_covid19_confirmed_US)
df_time_series_covid19_confirmed_US
```

```python
time_series_covid19_deaths_US = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv"
df_time_series_covid19_deaths_US = pd.read_csv(time_series_covid19_deaths_US)
df_time_series_covid19_deaths_US
```

## df_time_series_covid19_confirmed_US

```python
# Lets categorize per province_state
df_province = df_time_series_covid19_confirmed_US["Province_State"].to_frame().drop_duplicates()
df_province
```

```python
df_alabama = df_time_series_covid19_confirmed_US[df_time_series_covid19_confirmed_US["Province_State"]=="Alabama"]
df_alabama_autauga = df_alabama[df_alabama['Admin2']=="Autauga"]
df_alabama_autauga_to_graph = df_alabama_autauga.transpose().iloc[11:, :]

df_alabama_autauga_to_graph_new = df_alabama_autauga_to_graph.rename(columns={0: 'Confirmed Case Autauga'})
df_alabama_autauga_to_graph_new
```

```python
df_alabama_autauga_to_graph_new.plot()
plt.show()
```

## df_time_series_covid19_deaths_US

```python
# Lets categorize per province_state
df_province = df_time_series_covid19_deaths_US["Province_State"].to_frame().drop_duplicates()

df_alabama_deaths = df_time_series_covid19_deaths_US[df_time_series_covid19_deaths_US["Province_State"]=="Alabama"]
df_alabama_autauga_deaths = df_alabama_deaths[df_alabama_deaths['Admin2']=="Autauga"]
df_alabama_autauga_to_graph_deaths = df_alabama_autauga_deaths.transpose().iloc[12:, :]

df_alabama_autauga_to_graph_new_deaths = df_alabama_autauga_to_graph_deaths.rename(columns={0: 'Confirmed Case Autauga'})
df_alabama_autauga_to_graph_new_deaths
```

```python
df_alabama_autauga_to_graph_new_deaths.plot()
plt.show()
```
